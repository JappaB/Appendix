```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import itertools
import os

%matplotlib inline
```

```python
#Column names as 'global variables'
MY_POSITION = "My position"
STRATEGY_PARAMS = "Strategy Parameters"
TOURNAMENT_ID = "Tournament id"
DOMAIN = "Domain"
UTILITY = "Utility"
PROFILE = "Profile"
```

```python
class Session():
    
    def __init__(self, dataframe, position=None):
         
        self.data = dataframe
        self.strategy_parameters = parse_strategy_parameters(dataframe[STRATEGY_PARAMS].iloc[0])
        self.position = position
        
        if position:
            self.profile = self.get_label(PROFILE, select_me=True)
            self.domain = self.get_label(DOMAIN)
            
            # non-RLBOA agents don't have flag for their position, fill it manually
            if self.data[MY_POSITION].isnull().any():
                self.data[MY_POSITION] = position
                
    def get_label(self, col, select_me=False):
        '''Returns a single entry ot of -col-. Use -select_me- == True if there are seperate columns for
        agent 1 and agent 2'''
        
        if select_me:
            return self.data["{} {}".format(col, self.position)].iloc[0]
        
        return self.data[col].iloc[0]
    
    def get_column(self, col, select_me=False):
        '''Returns the values of column `col` that correspond to the agent of interest'''
        
        if select_me:
            header = f'{col} 1', f'{col} 2'
        else:
            header = f'{col}', f'{col}'

        
        pos_1_values = self.data.where(self.data[MY_POSITION] == 1, other=0.0)[header[0]]
        pos_2_values = self.data.where(self.data[MY_POSITION] == 2, other=0.0)[header[1]]
        
        return pos_1_values + pos_2_values
                
    def get_bins(self):
        return int(self.hyperparams['bins'])
    
    def get_domain(self):
        return self.domain
    
    def get_profile(self):
        
        PROFILES = {
            'adg_deal' : 'a',
            'adg_deal2' : 'b',
            'mary' : 'a',
            'sam' : 'b',
            'laptop_buyer_utility' : 'a',
            'laptop_seller_utility' : 'b',
            'Amsterdam_party1' : 'a',
            'Amsterdam_party2' : 'b',
            'energy_consumer' : 'a',
            'energy_producer' : 'b',
            'JaspAronLargeDiscreteDomain_util1' : 'a',
            'JaspAronLargeDiscreteDomain_util2' : 'b',
            
            'JaspAronMediumDiscreteDomain_util1' : 'a',
            'JaspAronMediumDiscreteDomain_util2' : 'b',
            
            'JaspAronSmallDiscreteDomain_util1' : 'a',
            'JaspAronSmallDiscreteDomain_util3' : 'b',
            
            'JaspAronLargeDiscreteDomain_util3' : 'a',
            'JaspAronLargeDiscreteDomain_util4' : 'b',
            
            'JaspAronMediumDiscreteDomain_util3' : 'a',
            'JaspAronMediumDiscreteDomain_util4' : 'b',
            
            'JaspAronSmallDiscreteDomain_util4' : 'a',
            'JaspAronSmallDiscreteDomain_util5' : 'b',
        }
        
        return PROFILES[self.profile.replace('.xml', '')]
    
    def get_opponent(self):
        return self.strategy_parameters['opponentName']
    
    def opponent_model(self):
        return self.strategy_parameters['opponentModel']
    
    def get_epsilon(self):
        return float(self.strategy_parameters['epsilon'])
    
    def get_utility(self):
        return self.get_column(UTILITY, select_me=True)
```

```python
def parse_strategy_parameters(full_string):
    
    full_string = str(full_string)
    pairs = full_string.split()

    hyperparams = [tuple(pair.split("=")) for pair in pairs]
    hyperparams = list(filter(lambda x: len(x) == 2, hyperparams))
    hyperparams = {key : value for key, value in hyperparams}
    
    try:
        hyperparams['bins'] = int(hyperparams.get('bins', None))
        hyperparams['epsilon'] = float(hyperparams.get('epsilon', None))
        hyperparams['alpha'] = float(hyperparams.get('alpha', None))
    except Exception as e:
        pass
        
    return hyperparams

def plot_data(ax, plot_data, xlabel, ylabel, label="", color="blue", pool = 20):
    
    # prepare data
    plot_data = np.array(plot_data)
    pooled_data = pool_sequence(plot_data, pool)
    steps = np.arange(plot_data.shape[0] / pool) * pool

    # plot in figure
    line = ax.plot(steps, pooled_data, label=label, color=color)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    
    return line[0]
        
def pool_sequence(plot_data, pool_size):
    pooled_data = []
    for i in range(0, plot_data.shape[0], pool_size):
        pooled_data.append(plot_data[i:i+pool_size].mean())
    
    return np.array(pooled_data)

def pick_data_with_params_and_domain(data, params, domains = [], profiles = []):

    result = []
    
    for session in data:
        matching_params = []
        
        for key in params.keys():
            matching_params.append(session.strategy_parameters[key] in params[key])
            
        correct_profile = session.get_profile() in profiles or profiles == []
        correct_domain = session.get_domain() in domains or domains == []
        
        if all(matching_params) and correct_domain and correct_profile:
            result.append(session)
            
    return result

def split_utility_profiles(data):
    
    position_1 = data.iloc[1::2]
    position_2 = data.iloc[0::2]
        
    return position_1, position_2
```

##  Experimental setup 1: Baseline experiments

```python
def parse_logs_default(path_to_log_files):

    log_files = []
    for file_name in os.listdir(path_to_log_files):
        if file_name[-3:] == 'csv':
            data = pd.read_csv(path_to_log_files + file_name, delimiter=';', header=1)
            data = split_utility_profiles(data)
            log_files.extend(data)
    
    sessions = []
    for i, data in enumerate(log_files):
        position = 1 if i % 2 == 0 else 2 # 1, 2, 1, 2, 1, 2
        sessions.append(Session(data, position))
    
    return sessions
```

### Usage:
Below are three cells that can be used to analyze the data of the baseline experiment.

The first cell  loads the data in session objects. Specify the relative path to the directory containing the logfiles in PATH_TO_LOGFILES. This can be either training logs or evaluation logs. You can select a specific subset of experiments by fillinf in the params, domains and/or profiles parameters in the __pick_data_with_params_and_domain__ function.

The second  cell plots the data. Use the POOL variable to POOL the data enabling smoother progress curves. the SLICE variable allows to select a certain slice of the data (start, stop, step). The plots are numbered and further information can be found in the output of the third cell at the corresponding number.

The third cell gives some further information per session:
- The xml file used
- The opponent
- The Average utility over the final 20 rounds
- The preference profile of the RL-agent

For the preference profiles it is important to not that the 2 lowest numberd preference profiles (1 and 2) always indicate a low opposition scenario together. The 2 higher numbered preference profiles (3 and 4 and in one case 4 and 5) form the high opposition scenarios. Also, of each preference profile combination (e.g. 1 and 2), the lowest always starts.

```python
PATH_TO_LOGFILES = '../experiments/lookback_scenarios/logs/'
individual_tournaments = parse_logs_default(PATH_TO_LOGFILES)
sessions = pick_data_with_params_and_domain(individual_tournaments, params = {
}, domains = [], profiles = [])
```

```python
POOL = 50
SLICE = slice(None, None, None)

for i, session in enumerate(sessions):
    fig, axes = plt.subplots(1,1)
    line1 = plot_data(axes, session.get_utility()[SLICE], xlabel="Negotiation Sessions", ylabel="Utility", color="tab:red", pool=POOL)
    axes.set_ylim(bottom=0, top=1)
    axes2 = axes.twinx()
    axes2.set_ylim(bottom=0, top=40)
    line2 = plot_data(axes2, session.get_column('Round')[SLICE], xlabel="Negotiation Sessions", ylabel="Rounds per session", pool=POOL)
    plt.legend((line1, line2), ("Utility", "Rounds"), loc='upper left')
    plt.title("Session: "+str(i))
```

```python
for i, session in enumerate(sessions):
    
    if session.profile[-5:] in (1,2):
        opposition = ' with low opposition'
    else:
        opposition = ' with high opposition'
        
    print("________")
    print("Session: " + str(i))
    print("Scenario: " + session.profile[8:14] +" domain" + opposition )
    util = session.get_column(UTILITY, select_me=True)
    print("profile: " +session.profile)
    print("Opponent: "+session.strategy_parameters['opponentName'])
    print("Mean utility over the last 20 training/eval steps: "+'{:.2f}'.format(session.get_column(UTILITY, select_me=True)[-20:].mean()))
    print("________")
    print()
```

## Comparison to RandomAgent
Below are three cells to compare the results of the agent with the RandomAgent: an agent with the same opponent model and acceptance strategy as the RL-agent, but a random bidding strategy. This can give some insight in whether the agent learned a non-trivial strategy.

### Usage
- Fill in the relative path to the random scenarios
- Run the two cells below. The RandomAgent is compared with the previously loaded negotiations from "PATH_TO_RANDOM_SCENARIO_LOGS". The output should be self-explanatory.

- The third cell shows the distribution of the random agent's utility and the mean of the RL-agent in that scenario

```python
PATH_TO_RANDOM_SCENARIO_LOGS = '../experiments/random_scenarios/logs/'
random_scenarios = parse_logs_default(PATH_TO_RANDOM_SCENARIO_LOGS)
```

```python
count = 0
for i in range(len(random_scenarios)):
    rand = random_scenarios[i].get_column(UTILITY, select_me=True)
    test = individual_tournaments[i].get_column(UTILITY, select_me=True)
    
    rand_mean_two_stdv =  rand.mean() + 2*rand.std()
    
    better = test[-20:].mean() - rand.mean() > 2*rand.std()
    if better:
        count += 1
        
    entry = individual_tournaments[i].get_opponent(), individual_tournaments[i].get_label(PROFILE, select_me=True)
    
    if entry[1][-5:] in (1,2):
        opposition = ' with low opposition'
    else:
        opposition = ' with high opposition'
    
    if entry[1][8:9] == "S":
        domain = "Small"
    
    elif entry[1][8:9] == "M":
        domain = "Medium"
    else:
        domain = "Large"
        
    
    print("________")
#     print(entry)
    print("Setting: " + domain +" domain" + opposition + " versus the " + entry[0] + " agent" )
    print()
    print("More than two standard deviations higher than the mean: ", better, f'\nRandom mean: {rand.mean():.2f}\nRL-agents\' average utility over last 20 training/eval steps: {test[-20:].mean():.2f}')
    print("Random mean + 2* std: "+str(rand_mean_two_stdv))
    print("________")
    print()

print("In total, " + str(count) + " of the " + str(len(random_scenarios)) + " negotiation settings above where more than two standard deviations higher than the mean ")
```

```python
for i in range(len(random_scenarios)):
    
    entry = individual_tournaments[i].get_opponent(), individual_tournaments[i].get_label(PROFILE, select_me=True)
    
    if entry[1][-5:] in (1,2):
        opposition = ' with low opposition'
    else:
        opposition = ' with high opposition'
    
    if entry[1][8:9] == "S":
        domain = "Small"
    elif entry[1][8:9] == "M":
        domain = "Medium"
    else:
        domain = "Large"
    
    plt.figure()
    util_random = random_scenarios[i].get_column(UTILITY, select_me=True)
    mean_test = individual_tournaments[i].get_column(UTILITY, select_me=True).mean()
    plt.hist(util_random)
    plt.axvline(mean_test, color='r')
    plt.title("Setting: " + domain +" domain" + opposition + " versus the " + entry[0] + " agent" )
```

## Experimental setup 2 & 3: Scenario and Opponent generality experiments

### Usage:
Regarding the usage there is just one difference with the cells above for experimental setup 1. There are no specific domains/scenarios for the training log files. So if you want to know the results for a specific domain/scenario, give the path to the logfile from an evaluation run and set the EVAL variable to True. Furthermore, specify wether the current analysis is a _"scenario"_ or an _"opponent"_ generality analysis in the ANALYSIS variable.



```python
def parse_logs_general(path_to_log_file):

    data = pd.read_csv(path_to_log_file, delimiter=';', header=1)
    return Session(data)
```
